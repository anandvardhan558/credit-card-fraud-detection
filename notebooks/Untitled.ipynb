## Imbalanced DataSet

#### UnderSampling:- In this we are randomly deleting rows from majority class to match them with minority class

```python
normal = df[df['Class'] == 0] 
fraud = df[df['Class'] == 1] 
```

```python
normal.shape
```

```python

fraud.shape
```

```python
normal_sample = normal.sample(n=473)
# This will select 473 random sample from normal transition(275190) to match fraud transition
```

```python
normal_sample.shape
```

```python
new_data = pd.concat([normal_sample, fraud], ignore_index = True)
```

```python
new_data['Class'].value_counts()
# Now we have same number of transition for both fraud and normal transation
```

```python
new_data.head()
```

```python
x = new_data.drop('Class', axis = 1)
y = new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

### After undersampling LogesticRegression is the best model, but in UnderSampling we lost a lot of valuable data. So we use OverSampling

 ### OverSampling using SMOT(Synthetic Minority Oversampling Technique) commonly used in Imbalances data. SMOT class distribution by randomly increasing minority class examples by replicating them. SMOT synthesis new minority instances between existing minority instances. SMOT generates the virtual training records by linear interpolation. Advantage of SMOT is that it doesnot generate duplicates but rather creating synthetic data points that are slightly different from the origional data points
